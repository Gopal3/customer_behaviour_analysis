```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
df
```

```python
df.columns = df.columns.str.lower()
```

```python
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_usd'})
```

```python
labels = ["Young", "Adult", "Middle-aged", "Senior"]
```

```python
df["age_group"] = pd.qcut(df["age"], q=4, labels = labels)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'We4ekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weeekly': 14,
    'Annually' : 365,
    'Every 3 Months': 90
    
}


df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
from sqlalchemy import create_engine
import urllib

database = "Customer_Shopping_Project"

params = urllib.parse.quote_plus(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=localhost\\MSSQLSERVER02;"
    f"DATABASE={database};"
    "Trusted_Connection=yes;"
    "Encrypt=no;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```

```python
table_name = "customer" 
df.to_sql(table_name, con=engine, if_exists='replace', index=False)

print(f"Data from CSV successfully loaded into table '{table_name}' in database '{database}'.")
```
